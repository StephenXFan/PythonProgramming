```python
import numpy as np
import cv2 as cv

from google.colab.patches import cv2_imshow

img = cv.imread('IMG_20141001_181249.jpg')
cv2_imshow(img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
cv.waitKey(0)
cv.destroyAllWindows()
```
