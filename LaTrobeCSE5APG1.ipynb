```python
more = True
while (more == True):
  print('*** Shape Maker ***')
  print()
  print('   1. Draw a triangle on the shape')
  print('   2. Draw a rectangle on the shape')
  print('   3. Exit')
  choice = int(input('Enter your choice: '))
  print()

  if (choice == 1):

    width_rectangle = int(input('Enter width of the rectangle: '))
    height_rectangle = int(input('Enter height of the rectangle: '))

    # smaller of width_rectangle ans height_rectangle
    smaller = width_rectangle
    if (height_rectangle < smaller):
      smaller = height_rectangle

    border = input('Do you want to draw the border (Y/N): ')

    # if there is a border
    if (border.upper() == 'Y'):
      max_triangle_height = smaller - 3
      triangle_height = int(input('Enter triangle height (min: 1, max: ' + str(max_triangle_height) + '): '))
      character_border = input('Enter a character to draw the border: ')

    # if there isn't a border
    elif (border.upper() == 'N'):
      max_triangle_height = smaller
      triangle_height = int(input('Enter triangle height (min: 1, max: ' + str(max_triangle_height) + '): '))
    
    character_shape = input('Enter a character to draw the inner content: ')

    print()

    # start from row 1
    m = 1

    # until last row
    while (m <= height_rectangle):

      # start from column 1
      n = 1

      # until last column
      while (n <= width_rectangle):

        # if there is a border
        if (border.upper() == 'Y'):

          # outside of the triangle border
          if (m > (triangle_height + 3) or n > (triangle_height + 3) or m < n):
            print('.', end = ' ')

          # inside of the triabgle border
          elif (m >= 3 and m <= (triangle_height + 2) and n >= 2 and n <= (triangle_height + 1) and m > n):
            print(character_shape, end = ' ')   

          # the triangle border
          else:
            print(character_border, end = ' ')  

        # if there isn't a border
        elif (border.upper() == 'N'):

          # outside of the triangle
          if (m > triangle_height or n > triangle_height or m < n):
            print('.', end = ' ')
            
          # the triabgle
          else:
            print(character_shape, end = ' ')  

        n = n + 1

      print()

      m = m + 1  

    print()

  
  elif (choice == 2):

    width_rectangle = int(input('Enter width of the rectangle: '))
    height_rectangle = int(input('Enter height of the rectangle: '))

    border = input('Border around the shape? (y/n): ')

    character_shape = input('Character to draw the shape: ')

    # if there is a border
    if (border.upper() == 'Y'):
      character_border = input('Character to draw the border: ')
      max_width_shape = width_rectangle - 2
      max_height_shape = height_rectangle - 2

    # if there isn't a border
    elif (border.upper() == 'N'):
      max_width_shape = width_rectangle
      max_height_shape = height_rectangle

    width_shape = int(input('Width of the shape (min: 1, max: ' + str(max_width_shape) + '): '))
    height_shape = int(input('Height of the shape (min: 1, max: ' + str(max_height_shape) + '): '))
    print()

    # start from row 1
    m = 1

    # until last row
    while (m <= height_rectangle):

      # start from column 1
      n = 1

      # until last column
      while (n <= width_rectangle):

        # if there is a border
        if (border.upper() == 'Y'):

          # outside of the shape border
          if (m > (height_shape + 2) or n > (width_shape + 2)):
            print('.', end = ' ')

          # inside of the shape border
          elif (m >= 2 and m <= (height_shape + 1) and n >= 2 and n <= (width_shape + 1)):
            print(character_shape, end = ' ')   

          # the shape border
          else:
            print(character_border, end = ' ')  

        # if there isn't a border
        elif (border.upper() == 'N'):

          # outside of the shape
          if (m > height_shape or n > width_shape):
            print('.', end = ' ')
            
          # the shape
          else:
            print(character_shape, end = ' ')  

        n = n + 1

      print()

      m = m + 1  

    print()

    
  elif (choice == 3):
    print('Quittingâ€”Done.')

    # exit the while loop
    more = False
```
