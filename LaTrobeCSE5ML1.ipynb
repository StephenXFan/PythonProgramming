## Task 1. Load the dataset, do basic data pre-processing, and split the dataset







### Load the dataset


```python
# Load the dataset using “pandas”,
# print the length of the dataset and display the first 5 rows of the dataset.

import pandas as pd

dataset = pd.read_csv("Life Expectancy.csv")

print("dataset length:", len(dataset))
dataset.head()
```

```python
# Check dataset shape
dataset.shape
```

### Deal with any missing values and use a proper method to handle categorical variables

```python
# Check if there are any missing values in the dataset.
dataset.isna().sum()
```

```python
# Since there are only very small number of rows with missing values in the dataset,
# those rows are just removed for easy processing.
dataset = dataset.dropna()
print("dataset length:", len(dataset))
```

```python
# Check if there are any duplicated rows.
dataset.duplicated().any()
```

```python
# Get a summary of numerical columns.
dataset.describe()
```

```python
# Check value count for the categorical variable "Status".
print(dataset.Status.value_counts(),"\n")
```

```python
# Transform the nominal variable "Status" into dummy variables “Status_Developing” and “Status_Developed”.
dataset = pd.get_dummies(dataset, columns=['Status'])

# Drop the "Status_Developed" column
dataset.drop(['Status_Developed'], axis = 1, inplace = True)

dataset.head()
```

```python
# Check dataset shape
dataset.shape
```

### Define the input variables and the target variable


```python
# The first column (with column index 0) is the target variable "Life expectancy".
# All other columns (with column index from 1 to the end) are the input variables.

array = dataset.values
X = array[:,1:]
y = array[:,0]
```

### Split the dataset into training and testing (with 10% of the dataset for testing)


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

### Apply normalisation on X (both the training and test set)


```python
from sklearn.preprocessing import MinMaxScaler

# Fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# Transform training data
X_train_norm = norm.transform(X_train)

# Transform testing data
X_test_norm = norm.transform(X_test)
```

## Task 2. Train and evaluate the two regression models on the training set with a cross-validation method, optimise the models, and evaluate models on the test set



### Define the two regression models, including Linear Regression and SVM, with their default settings

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
from sklearn.svm import SVR

svr = SVR()
```

### Define a 5-fold cross validation to train and evaluate the two models based on the average score


####  1) Define a 5-fold cross validation with data shufflling and setting the random state to 123


```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=123)
```

#### 2) Run a 5-fold cross validation and print the average r-squared scores based on the cross validation results




1. For the Linear Regression model

```python
# Basic training of the Linear Regression model

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

# Define a Linear Regression model with default parameter setting
lr = LinearRegression()

# Run the previously defined 5-fold cross validation on the dataset
results = cross_val_score(lr, X_train_norm, y_train, cv=kfold)

# Print the average r-squared score
print("Average R2 of LR:", results.mean())
```

```python
results
```

2. For the SVM model

```python
# Basic training of the SVM model

from sklearn.svm import SVR

svr = SVR()
results = cross_val_score(svr, X_train_norm, y_train, cv=kfold)
print("Average R2 of SVM:", results.mean())
```

```python
results
```

### Apply parameter finetuning steps to the two models separately to optimise the model performances and compare the cross-validated results before and after finetuning for each model

1. Optimise the Liner Regression model

```python
# Finetune parameters for the Linear Regression model

from sklearn.model_selection import GridSearchCV

grid_params_lr = {
    'fit_intercept': [True, False],
    'positive':      [True, False]
}

lr = LinearRegression()
gs_lr_result = GridSearchCV(lr, grid_params_lr, cv=kfold).fit(X_train_norm, y_train)
print(gs_lr_result.best_score_)
```

2. Optimise the SVM model

```python
# Finetune parameters for the SVM model

from sklearn.model_selection import GridSearchCV

grid_params_svr = {
    'kernel': ('linear', 'poly', 'rbf', 'sigmoid'),
    'C':      [10, 50],
    'degree': [2, 3],
    'coef0':  [0.001, 0.005],
    'gamma':  ('auto', 'scale')
}

svr = SVR()
gs_svr_result = GridSearchCV(svr, grid_params_svr, cv=kfold).fit(X_train_norm, y_train)
print(gs_svr_result.best_score_)
```

### Evaluate the two optimised models (with the best parameter setting from the above step for each model type) on the testing dataset, then compare the results with what you got from before finetuning



1. Evaluate the trained Linear Regression model using testing dataset

```python
# Use the best model and evaluate on testing dataset
lr_test_R2 = gs_lr_result.best_estimator_.score(X_test_norm, y_test)
print("R2 of LR in testing:", lr_test_R2)
```

```python
# Check the parameter setting for the best selected model
gs_lr_result.best_params_
```

2. Evaluate the trained SVM model using testing dataset

```python
# Use the best model and evaluate on testing dataset
svr_test_R2 = gs_svr_result.best_estimator_.score(X_test_norm, y_test)
print("R2 of SVM in testing:", svr_test_R2)
```
